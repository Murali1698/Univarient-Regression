```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def RELU(x):
 x1=[]
 for i in x:
  if i<0:
   x1.append(0)
  else:
   x1.append(i)
 return x1
```

```python
x = np.linspace(-20,10,100)
print(x)
```

```python
print(RELU(x))
```

```python
print(x,RELU(x))
```

```python
plt.plot(x, RELU(x))
plt.axis('tight')
plt.title('Activation Function :ReLU')
plt.show()
```
