```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
 return 1/(1+np.exp(-x))
```

```python
x = np.linspace(-10, 10,100)
print(x)
```

```python
print(sigmoid(x))
```

```python
print(x,sigmoid(x))
```

```python
plt.plot(x, sigmoid(x))
plt.axis('tight')
plt.title('Activation Function :Sigmoid')
plt.show()
```
