```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def softmax(x):
 return np.exp(x) / np.sum(np.exp(x), axis=0)
```

```python
x = np.linspace(-20,20,100)
print(x)
```

```python
print(softmax(x))
```

```python
print(x,softmax(x))
```

```python
plt.plot(x, softmax(x))
plt.axis('tight')
plt.title('Activation Function :softmax')
plt.show()
```
